## Data Communications and Concepts

### Fundamentals of Storytelling

- To make results stick, they should be:
1) Simple
2) Concrete
3) Credible

#### Why are Stories needed?
- The best results have no impact without proper presentation
- The aim is to convinve change adverse stakeholders to change
- These may be non technical stakeholders who might not be comfortable with maths and statistics

#### Data Storytelling:
- Data storytelling is the practice of building a narative around a set of data and its accompanying visualizations to help convey the meaning of that data in a powerful and compelling fashion

1) 3 minute story:
- What would you say in 3 minutes
2) Big Idea:
- Unique point of view in one sentence



#### Techincal or non Technical Approach

- Assess the audience on how technical they may be
- Data Professionals care about their methods 
- Audience likely cares more about the results and its implications


### ADEPT Technique

- A -> Analogy
- D -> Diagram
- E -> Example
- P -> Plain English
- T -> Technical Definition

### Compelling Narative

- Meaningful to target audience
- Prioritize key points
- Drive Change

### Narative Structure:



1) Background:
- What motivated the analysis?
- What changed?
- Who is the focus of the analysis?

2) Insight 1:
- What contributed to the problem?
- Only mention relevant information

3) Insight 2:
- Add supporting evidence
- only if it helpsbetter explain the cause of the problem

4) Climax:
- Central finding
- What would happen if there is no change

5) Next Steps:
- Potential Solutions
- Course of action
- Proactive

### Building Narative:

We can focus on the following parameters:

1) Change over time: 
- Change in data over seasons, etc

2) Correlation:
look at the correlations between the data

3) Comparision:

4) Clustering:
