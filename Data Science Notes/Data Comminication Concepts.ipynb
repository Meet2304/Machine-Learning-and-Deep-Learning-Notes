## Data Communications and Concepts

### Fundamentals of Storytelling

- To make results stick, they should be:
1) Simple
2) Concrete
3) Credible

#### Why are Stories needed?
- The best results have no impact without proper presentation
- The aim is to convinve change adverse stakeholders to change
- These may be non technical stakeholders who might not be comfortable with maths and statistics

#### Data Storytelling:
- Data storytelling is the practice of building a narative around a set of data and its accompanying visualizations to help convey the meaning of that data in a powerful and compelling fashion

1) 3 minute story:
- What would you say in 3 minutes
2) Big Idea:
- Unique point of view in one sentence



#### Techincal or non Technical Approach

- Assess the audience on how technical they may be
- Data Professionals care about their methods 
- Audience likely cares more about the results and its implications


### ADEPT Technique

- A -> Analogy

### Compelling Narative

- Meaningful to target audience
- Prioritize key points
- Drive Change

### Narative Structure:



1) Background:
- What motivated the analysis?
- What changed?
- Who is the focus of the analysis?

2) Insight 1:
- What contributed to the problem?
- Only mention relevant information

3) Insight 2:
- Add supporting evidence
- only if it helpsbetter explain the cause of the problem

4) Climax:
- Central finding
- What would happen if there is no change

5) Next Steps:
- Potential Solutions
- Course of action
- Proactive

### Reports

#### Types of Reports:

1) Informational Reports:
- Factual Information
- Short
- Not Strict Structure
- Inform about facts

2) Analytical Reports:
- are used to present an analysis
- maybe short or long
- Have a strict structure
- Purpose is to drive descisions based on data

3) Final Reports:
- Includes data analysis, findings and results and visuals
- They are long

4) Summary Reports:
- Only have key findings and recommendations
- Use visuals
- Are usually short
- Link to the main document
- made for descisino makers that do not need technical details





### Report Structure

#### -> Report Structure for data reports in Journals:

1) Introduction:
- Purpose of the report 
- Contextual information about why we performed the analysis
- Ask which factors are responsible for the problem

2) Body:
- Data
- Methods used to analyse data
- Analysis of the data
- Results of the analysis

3) Conclusions:
- Restate the questions
- Summarize important results
- Provide recommendations

#### -> Report Structure for data reports in business context:

1-3-25 approach

- 1 page of abstract
- 3 pages of executive summary
- 25 pages of detail

### Reproducibility:

- the ability to obtain similar results using the same general approach, but a different environment (the data is processed in the same way, but using a different language on a different OS).


- Reports should be concise and precise. Avoid misleading and confusing facts

### Oral Presentations:

#### Points to Consider

1) Purpose of the presentation
- Consider the type of presentation we are giving.

2) Audience 
- How big is the audience

3) Message are trying to convey
- What is the central message (One Sentence)
- Audience will forget 90% of the presentation after one week
- Closing statement should sum up the presentation and strengthen the central message

4) How long do you have?

#### Structure of an Oral Presentation

1) Introduction
2) Methods, analysis and model outputs
3) Conclusions and takeaways
