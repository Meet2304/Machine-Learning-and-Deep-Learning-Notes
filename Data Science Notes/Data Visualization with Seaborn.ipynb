### Relational Plots and Subplots

RELATIONAL PLOT
- Scatterplots or lineplots
- lets us create subplots in a single figure
Using the col parameter


Using the row parameter


Using both row and col parameters


#### Wrapping Coulmns



```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.scatterplot(x = "total_bill", y = "tip", data = tips)
plt.show()


# USING REPLOT
# Using the col parameter
sns.relplot(x = "total_bill", y = "tip", data = tips, kind = "scatter", col = "smoker")
plt.show()
#or
# Using the row parameter
sns.relplot(x = "total_bill", y = "tip", data = tips, kind = "scatter", row = "smoker")
plt.show()
#or
# Using both the row and col parameters
sns.relplot(x = "total_bill", y = "tip", data = tips, kind = "scatter", col = "smoker", row = "time")
plt.show()

# WRAPPING COLUMNS
sns.relplot(x = "total_bill", y = "tip", data = tips, kind = "scatter", col = "day", col_wrap = 2)
plt.show()

# Ordering Columns
sns.relplot(x = "total_bill", y = "tip", data = tips, kind = "scatter", col = "day", col_wrap = 2, col_order = ["Thur", "Fri", "Sat", "Sun"])
plt.show()
```

### Customizing Scatter Plots

- Scatter plots show the relationship between two quantitative variables

```python
import seaborn as sns
import matplotlib.pyplot as plt

# SUBGROUPS WITH POINT SIZE
sns.relplot(x = "total_bill", y = "tip", data = tips, kind = "scatter", size = "size", hue = "size")
plt.show()

# SUBGROUPS WITH POINT STYLE
sns.relplot(x = "total_bill", y = "tip", data = tips, kind = "scatter", hue = "smoker", style = "smoker")
plt.show()


# CHANGING POINT TRANSPARENCY
sns.relplot(x = "total_bill", y = "tip", data = tips, kind = "scatter", alpha = 0.4)
plt.show()
```

### Box Plot

- Used to compare the distribution of a quantative variable across different groups of a categorical variable
- By default the whiskers extend to 1.5 times the inter quartile range
- IQR = 25th percentile to 75th percentile

```python
import seaborn as sns
import matplotlib.pyplot as plt

g = sns.catplot(x = "time", y = "total_bill", data = tips, kind = "box")
plt.show()

# changing order of categories
g = sns.catplot(x = "time", y = "total_bill", data = tips, kind = "box", order = ["Dinner", "Lunch"])
plt.show()

# Omitting Outliers
# 'sym" parameter is used with the boxplot function and showfliers parameter is used with the catplot function
# "Sym" parameter is used to change the apperance of the outliers and omit the outliers when it is left empty
g = sns.catplot(x = "time", y = "total_bill", data = tips, kind = "box", showfliers = False)
plt.show()

# Changing the way the whiiskers are defined 
# makes them extend to 2*IQR by whis = 2.0
# show the 5th and 95th percentiles by whis = [5, 95]
# show the min and max values by whis = [0, 100] but there will be no outliers in this case as the graph covers the entirity of the data
```

### Point Plots

- Show the mean of quantitative variable for the observations in each category
- Vertical lines show 95% confidence intervals




```python
sns.catplot(x = "age", y = "masculinity_important", data = masculinity_data, hue = "feel_masculine", kind = "point")
plt.show()

# Removing the lines connecting each point
sns.catplot(x = "age", y = "masculinity_important", data = masculinity_data, hue = "feel_masculine", kind = "point", join = False)
plt.show()
```

```python
from numpy import median
sns.catplot(x = "smoker", y = "total_bill", data = tips, kind = "point")
plt.show()

# To calculate ci for median instead of the mean
# Useful if the dataset has a lot of outliers
sns.catplot(x = "smoker", y = "total_bill", data = tips, kind = "point", estimator = median)
plt.show()

# Setting capsize
sns.catplot(x = "smoker", y = "total_bill", data = tips, kind = "point", estimator = median, capsize = 0.2)
plt.show()


```
