### Scope



### Nested Function



```python
# Nested Functions

def raise_val(n):
    """Return the inner function"""
    def inner(x):
        """Raise x to the power of n"""
        raised = x ** n
        return raised
    return inner

square = raise_val(2)
cube = raise_val(3)
print(square(2), cube(4))
```

### Using Nonlocal

```python
def outer():
    """Prints the values of n"""
    n = 1

    def inner():
        nonlocal n
        n = 2
        print(n)
    inner()
    print(n)

outer()
```

### Default Arguments and Flexible Arguments

```python
# DEFAULTS ARGUMENTS

def power(number, pow = 1):
    """Raise number to the power of pow"""
    new_value = number ** pow
    return new_value

print(power(9, 2))
print(power(9))
```

### Flexible Arguments

- *ARGS:
 - converts the list of arguments into a tupple
- **kwargs:
 - converts the list of arguments and identifiers into a dictionary
 

```python
# FLEXIBLE ARGUMENTS

def add_all(*args):
    """Sum all values in args together"""
    sum_all = 0
    for num in args:
        sum_all += num
    return sum_all

add_all(5, 4, 2.0, 1) 
```

```python
def print_all(**kwargs):
    """Print out key-value pairs in **kwargs"""

    # Print out the key-value pairs
    for key, value in kwargs.items():
        print(key + ": " + value)

print_all(name = "dumbledore", job = "headmaster")
```

### Lambda Functions:

-Quicker way to write functions


```python
# LAMBDA FUNCTION
raise_to_power = lambda x, y: x**y

raise_to_power(2, 3)
```

```python
# Function map takes two arguments: map(func, seq)
# map() applies the function to all the elements in the sequence

nums = [48, 6, 9, 21, 1]
square_all = map(lambda num: num**2, nums)
print(square_all)
print(list(square_all))
```

### Error Handling



```python
# ERROR HANDLING
# This function works well with integers but throws errors when it is used with strings
def sqrt(x):
    """Returns the square root of a number"""
    return x ** (0.5)

sqrt(4)

```

```python
# TRY EXCEPT BLOCK

def sqrt(x):
    """ Returns the square root of a number """
    try:
        return x ** 0.5
    except:
        print("x must be an integer or a float")

sqrt('hello')
```

```python
# CATHCHING ONLY TYPEERRORS

def sqrt(x):
    """ Returns the square root of a number """
    try:
        return x ** 0.5
    except TypeError:
        print("x must be an integer or a float")

sqrt('hello')
```

```python
# RAISING ERRORS

def sqrt(x):
    """ Returns the square root of a number """
    if(x < 0):
        raise ValueError('x must be non negative integer')
    try:
        return x ** 0.5
    except TypeError:
        print("x must be an integer or a float")

sqrt(-2)
```
