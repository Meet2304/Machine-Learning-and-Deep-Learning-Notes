# Sampling in Python

### Sampling and Point Estimates

- Working with a subset of the whole dataset is called sampling
- Sample is the subset of data that we calculate on

- Pandas ".sample" method returns a radndom subset of rows

```python
# Importing the dataset
import pandas as pd
coffee_ratings = pd.read_feather('Datasets/coffee_ratings_full.feather')
coffee_ratings.head()
```

```python
# The relationship between cup points and flavour
pts_vs_flavor_pop = coffee_ratings[["total_cup_points", "flavor"]]

pts_vs_flavor_samp = pts_vs_flavor_pop.sample(10)
print(pts_vs_flavor_samp)
```

#### Population Parameter:
- Calculation made on the population dataset

#### Point Estimate (Sample Statistic):
- Calculation made on the sample dataset


### Convenience Sampling



- The data was collected by the easiest method - telephoning people. 
- As most of the people having telephones were relativly rich, the collected data did not accurately refelct the population correctly
- Histograms are a great way to visualize selection bias


```python
coffee_ratings["total_cup_points"].mean()
```

```python
# The convenienet sample is not representative  of the entire dataset

coffee_ratings_first10 = coffee_ratings.head(10)
coffee_ratings_first10["total_cup_points"].mean()
```

```python
# Visualising Selection bias by plotting histograms
import matplotlib.pyplot as plt
import numpy as np
coffee_ratings["total_cup_points"].hist(bins = np.arange(59, 93, 2))
plt.show()
coffee_ratings_first10["total_cup_points"].hist(bins = np.arange(59, 93, 2))
plt.show()
```

### Pseudo Random Number Generation

- Definition of Random: made, done, happening or chosen withoud method or conscious descision
- Pseudo Random values are called so because each random value is calculated from the previous random value. 
- Since we have to start the calculation somewhere, the first random value is generated from what we call the "seed" number.
- The same seed value yields the same random numbers





```python
# Visualising Random Numbers

randoms = np.random.beta(a = 2, b = 2, size = 5000)

plt.hist(randoms, bins = np.arange(0, 1, 0.05))
plt.show()
```

```python
np.random.seed(20000229)
# loc and scale argument represent the mean and the standard deviation
print(np.random.normal(loc = 2, scale = 1.5, size = 2))
print(np.random.normal(loc = 2, scale = 1.5, size = 2))
np.random.seed(20041004)
print(np.random.normal(loc = 2, scale = 1.5, size = 2))
print(np.random.normal(loc = 2, scale = 1.5, size = 2))
```
