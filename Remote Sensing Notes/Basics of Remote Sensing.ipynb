# FUNDAMENTALS OF REMOTE SENSING:

- Remote sensing is obtaining information about an object from a distance.
- Satellites carry instruments or sensors that measure electromagnetic radiation coming from the earth atmosphere system
- Electromagnetic Radiation: The energy Earth recieves from the Sun is called electromagnetic radiation.
- This electromagnetic radiation is reflected, absorbed and emitted by the earths atmosphere or surface.

# Spectral Signatures:

- Different materials absorb and reflect different components of the electromagnetic radiation.
- So, we can look at the reflected wavelengths detected by a sensor and determine the material it is made of.
- This is known as a spectral signature





## Interaction with the Earth's Surface:

1) Vegetation:

- Healthy vegetation absorbs blue and red but reflects green and far infrared

2) Water:

- Longer visible wavelengths(green and red) and near IR are absorbed more by water than shorter wavelenghts(blue)

3) Atmosphere:

- Satellites can provide information about the composition of the atmosphere by using the way the electromagnetic radiation interacts with the atmosphere
- This includes the presence of particulate matter which contributes to air pollution, clouds and precipitation.

## How do Satellites Collect Data?

- Solar Radiation passes through the atmosphere and hits a target surface like a forest, water, etc.
- Different materials reflect, emit and absorb at different wavelengths

1) Forests and Trees:

- Vegetation reflects certain wavelengths of solar radiation back to a satellite's sensors. Changes in the wavelengths reflected can lead to information on any land cover changes that might be occurring. 

2) Water:

- Water reflects radiation mostly in the blue-green spectrum. Changes in these reflected wavelengths can provide data on change in water levels, whether or not there is algal growth on the water, and more. 

3) Grass and Crops:

- Grass, crops, and bare soil reflect different types of radiation to a satellite's sensors.

4) Pavement and Buildings:

- Pavement and buildings both reflect different wavelengths of light than natural materials. This allows a satellite to collect data on the amount of development in an area. 

5) Atmosphere:

- Some of the solar radiation interacts with the atmosphere. When a satellite detects this radiation, it can evaluate what might have caused interference in the atmosphere.

## Characteristics of Satellites:

 Orbits: Low-Earth Orbit vs. Geostationary

 Energy Source: Passive vs. Active

 Solar and Terrestrial Spectra: Visible, UV, IR, Microwave...

 Measurement Technique: Scanning; Non-Scanning; Imager; Sounders

 Resolution Type and Quality: Spatial, Temporal, Spectral, Radiometric

 Application: Weather, Ocean Color, Land Mapping, Air Quality, Radiation Budget, etc.


### Types of Satellite Orbits:

1) Geostationary:

- Satellites are typically ~36,000 km over the equator with the same rotation period as Earth
- Multiple observations per day 
- Limited spatial coverage - observations are always of the same area
- Examples: Weather or communications satellites

2) Low Earth Orbits:

- Circular orbit moving relative to Earth at 160-2,000 km 
- can be polar or non-polar
- Less frequent measurements
- Large (global) spatial coverage
- Examples (polar): Landsat or Terra

#### a) Polar Low Earth Orbit:
 - Provides global coverage and measurement frequency can vary from 1 measurement per day to 1 per month.
#### b) Non-Polar Low Earth Orbit:
 - Does not provide global coverage. The measurement frequency can vary every few hours to a few weeks.

### Passive vs Active Sensors:

1) Passive Remote Sensing:
- depends on changes in gravity or reflected and emitted radiation from the Earth. For instance, the Grace Follow-On satellites measure changes in the Earths gravity. Landsat, the MODIS sensor onboard Terra & Aqua, and Aquas AIRS sensor, all depend on changes in the reflected and emitted radiation from the Earth for data.

2) Active Remote Sensing:
- an instrument sends beams of radiation and measures its return.
Examples: Radar, LIDAR

### Types of Resolutions:

1) Spatial Resolution:

- It is the geographical area covered by a pixel in a satellite's image.
- Higher the Spatial Resolution, the less area is covered in a pixel. 
- It is the measure of the smallest object that can be resolved by the sensor

2) Temporal Resolution:

- The time it takes for a satellite to image the same area at the same viewing angle a second time
- It depends on the satellite and sensor capabilities, the overlaps of the sensors path and the latitude of the satellite



3) Spectral Resolution:

- The abilility of the sensor to detect fine wavelength intervals.
- Instruments detect different ranges of wavelengths along with electromagnetic radiation
- The narrower a wavelength is the finer the spectral resolution is.

4) Radiometric Resolution:

- A sensors ability to discriminate the differences in energy.
- chief factor that determines it is saturation radiance-the brightest things the detectors can measure and how radiance measurments are turned into a digital bit stream. 

### Satellite Data Processing Levels:

- Level 0 & 1 is the raw instrument data that may be time-referenced. It is the most difficult to use. 
- Level 2 is Level 1 data that has been converted into a geophysical quantity through a computer algorithm (known as retrieval). This data is geo-referenced and calibrated.This data has the best spatial and temporal resolution 
- Level 3 is Level 2 data that has been mapped on a uniform space-time grid and quality controlled. 
- Level 4  is Level 3 data that has been combined with models or other instrument data.

-Level 3 &  4 data is the easiest to use.

### Advantages of Remote Sensing:

- Provides information where there are no ground based measurments
- Provides Globally consistent observations
- Can provide direstion on where to place ground based sensors
- Provides continuous monitoring of our planet

### Disadvantages of Remote Sensing:

- It is very difficult to obtain high spectral, spatial, temporal, and radiometric resolution all at the same time:
Satellites with low spatial resolution tend to have high temporal resolution, and satellites with high spatial resolution tend to have low temporal resolution
- Large amounts of data in a variety of formats can lead to more time and processing
- Applying satellite data may require additional processing, visualization, and other tools
